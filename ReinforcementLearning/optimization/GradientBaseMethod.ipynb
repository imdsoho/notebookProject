```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('ex1data1.txt', sep=',', header=None)
X = data[0]
Y = data[1]
```

```python
w = np.random.uniform(low=0.0, high=1.0, size=X.shape[0])
b = np.random.uniform(low=0.0, high=1.0, size=X.shape[0])
```

```python
#step_size = 0.1     
step_size = 0.01
#step_size = 0.001
#step_size = 0.0001
#step_size = 0.00001

ITERATION_COUNT = 5000
```

```python
def plot_show(X, Y, w, b):
    plt.figure(figsize=(16,8))
    plt.scatter(X, Y, color='blue') 
    Y_pred = w * X + b
    plt.plot(X, Y_pred, color='red')
    plt.show()
```

```python
data_count = X.shape[0]

for epoch in range(ITERATION_COUNT):
    y_hat = w * X + b

    # 목적 함수 
    error = (1/(2*data_count)) * np.sum( (y_hat - Y)**2 )
    
    # 멈춤 조건
    if error < 0.001:
        break
    
    # Decision variable theta update 규칙
    temp_0 = (1/data_count) * np.sum((y_hat - Y) * X)
    temp_1 = (1/data_count) * np.sum(y_hat - Y)
    
    w = w - step_size * temp_0
    b = b - step_size * temp_1
        
    # 1000번마다 결과 출력
    if epoch % 1000 == 0:
        print(f"Iteration {epoch}: Loss {error:.4f}, w: {w[0]:.4f}, b: {b[0]:.4f}")
        #plot_show(X, Y, w, b)
        
print(f"\n최종 결과 -> w: {w[0]:.4f}, b: {b[0]:.4f}")
```
