```python
import sys
import gurobipy as gp
from gurobipy import GRB
```

```python
m = gp.Model("lp1")
```

```python
meter = m.addVar(name="meter", vtype=GRB.INTEGER, lb=-GRB.INFINITY, ub=GRB.INFINITY)
```

```python
END_LINE = 100
```

```python
m.setObjective(meter/2, GRB.MINIMIZE)
```

```python
#m.addConstr(meter == END_LINE/2, name="c1")
#m.addConstr(meter >= 0 , name="c1")
```

```python
m.optimize()
```

```python
if m.status == GRB.OPTIMAL:
    print('Optimal solution found')
elif m.status == GRB.UNBOUNDED:
    print('Unbounded solution found')
elif m.status == GRB.INFEASIBLE:
    print('Infeasible solution found')
else:
    print('Optimal solution Not found')

for v in m.getVars():
    print('%s %g' % (v.varName, v.x))

print('Obj: %g' % m.objVal)
```
