```python
import sys
import gurobipy as gp
from gurobipy import GRB
```

```python
m = gp.Model("lp1")
```

```python
sa_buy_barrel = m.addVar(name="sa_buy_barrel", vtype=GRB.INTEGER, lb=0)
ve_buy_barrel = m.addVar(name="ve_buy_barrel", vtype=GRB.INTEGER, lb=0)
```

```python
SA_MONEY_PER_BARREL = 20            # 배럴 당 금액
VE_MONEY_PER_BARREL = 15

SA_BUY_AMOUNT_PER_DAY = 9000        # 구매 최대량
VE_BUY_AMOUNT_PER_DAY = 6000

SA_GAS = 0.3                        # 원유 배럴당 생산할 수 있는 제품 양
SA_JET = 0.4
SA_LUB = 0.2
SA_TSH = 0.1

VE_GAS = 0.4
VE_JET = 0.2
VE_LUB = 0.3
VE_TSH = 0.1

GOAL_GAS = 2000                     # 납품 해야 할 제품의 양
GOAL_JET = 1500
GOAL_LUB = 500
```

```python
m.setObjective(sa_buy_barrel*SA_MONEY_PER_BARREL + ve_buy_barrel*VE_MONEY_PER_BARREL, GRB.MINIMIZE)
```

```python
m.addConstr(sa_buy_barrel >= 0, name="c1")
m.addConstr(ve_buy_barrel >= 0, name="c2")

m.addConstr(sa_buy_barrel * SA_GAS + ve_buy_barrel * VE_GAS >= GOAL_GAS, name="c3")
m.addConstr(sa_buy_barrel * SA_JET + ve_buy_barrel * VE_JET >= GOAL_JET, name="c4")
m.addConstr(sa_buy_barrel * SA_LUB + ve_buy_barrel * VE_LUB >= GOAL_LUB, name="c5")

m.addConstr(sa_buy_barrel <= SA_BUY_AMOUNT_PER_DAY, name="c6")      # 구매 최대량 제약
m.addConstr(ve_buy_barrel  <= VE_BUY_AMOUNT_PER_DAY, name="c7")
```

```python
m.optimize()
```

```python
if m.status == GRB.OPTIMAL:
    print('Optimal solution found')
else:
    print('Optimal solution Not found')

for v in m.getVars():
    print('%s %g' % (v.varName, v.x))

print('Obj: %g' % m.objVal)
```
